## Transforming label column to numerical (binary) representation

```python
data['Label'] = data['Label'].apply(lambda label: 1 if label == 's' else 0)
```

## Plotting hists (s - red, b - blue)

```python
def plot_labeled_hist(data, column):
    fig, ax = plt.subplots()
    
    ax.hist(data[(data['Label'] == 1) & (data[column] != -999)][column], bins=100, color='Red', alpha=0.3)
    ax.hist(data[(data['Label'] == 0) & (data[column] != -999)][column], bins=100, color='Blue', alpha=0.3)
    
    ax.set_title(column)
```

```python
for column in data.columns[1:-2]:
    plot_labeled_hist(data, column)
```

## Plotting graphs Feature vs. Label

```python
def plot_labeled_scatter(data, column):
    fig, ax = plt.subplots()
    
    ax.scatter(data[(data['Label'] == 1) & (data[column] != -999)]['Label'], data[(data['Label'] == 1) & (data[column] != -999)][column], color='Red')
    ax.scatter(data[(data['Label'] == 0) & (data[column] != -999)]['Label'], data[(data['Label'] == 0) & (data[column] != -999)][column], color='Blue')
    
    ax.set_title(column)
```

```python
for column in data.columns[1:-2]:
    plot_labeled_scatter(data, column)
```

## Labels' balance

```python
s_count = len(data[data['Label'] == 1])
print 'Number of s - labels: %d (%.2f%%)' % (s_count, s_count / len(data) * 100)

b_count = len(data[data['Label'] == 0])
print 'Number of b - labels: %d (%.2f%%)' % (b_count, b_count / len(data) * 100)
```

We can see that there are twice more b-labels than s-labels. So we must use some methods to get rig of this class inequality.

## Exploring NaNs (-999.0)

```python
for column in data.columns[1:-2]:
    nan_count = len(data[data[column] == -999])
    print '%s: %d (%.2f%%)' % (column, nan_count, nan_count / len(data) * 100)
```

There are some features with many missing values.
It's interesting to note, that features like 'PRI_jet_subleading_pt', 'PRI_jet_subleading_eta', 'PRI_jet_subleading_phi', etc., have absolutely the same percentage of missing values (70.98%) (that's why they are so correlated).

There are many ways to deal with missing values like replacing them with feature's mean or median value, or we can just drop features whose percentage of missing values is bigger than X.XX%.

Let's look at correlations of ('DER_deltaeta_jet_jet' vs. 'PRI_jet_subleading_pt', 'PRI_jet_subleading_eta', 'PRI_jet_subleading_phi') without missing values:

```python
data[data['DER_deltaeta_jet_jet'] != -999][['DER_deltaeta_jet_jet', 'PRI_jet_subleading_pt', 'PRI_jet_subleading_eta', 'PRI_jet_subleading_phi']].corr()
```

So now we can see that correlation was caused by missing values

```python
def plot_weight_scatter(data, column):
    fig, ax = plt.subplots()
    
    ax.scatter(data[(data['Label'] == 1) & (data[column] != -999)][column], data[(data['Label'] == 1) & (data[column] != -999)]['Weight'], color='Red')
    ax.scatter(data[(data['Label'] == 0) & (data[column] != -999)][column], data[(data['Label'] == 0) & (data[column] != -999)]['Weight'], color='Blue')
    
    ax.set_title(column)
    ax.set_xlabel(column)
    ax.set_ylabel('Weight')
```

As we can see that using 'Weight' we can perfectly mark the data. So normed (!) 'Weight' column can be considered as probability to get to the b-label (maybe). Using it we can reduce the problem of classification to the regression problem.
